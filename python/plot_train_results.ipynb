```python
import matplotlib.pyplot as plt

def line_plot(data): 
    x = np.arange(1, data.size+1)
    m, b = np.polyfit(x, data, 1)


    plt.title("Training Score History")
    plt.xlabel("X axis")
    plt.ylabel("Y axis")
    plt.plot(x, data, color ="green")
    # Add linear regression line to scatterplot 
    plt.plot(x, m*x+b, color="red")
    plt.show()

    print(f'Line: y = {m}x + {b}')
```

```python
import numpy as np

data = np.loadtxt('train_results/score-history-2.csv', delimiter=',')
line_plot(data)
```
